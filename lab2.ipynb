```python
#Is the given list divisible by 3 or NOT?
age_tenants = [20, 22, 33, 24, 27, 21, 33, 20]

for x in age_tenants:
    if x%3 == 0:
        print(x)
    else:
        print("List is not divisible by 3")
        break
        
    
```

```python
#Square of even numbers in a list
for x in age_tenants:
    if x % 2 == 0:
       square=x**2
       print(square)

```

```python
#Sum of digits of all EVEN numbers in a list
age_tenants = [20, 22, 33, 24, 27, 21, 33, 20]
count=0

for x in age_tenants:
    if x % 2 == 0:
       count+=x
print("Sum of the even numbers are : ",count)
      
```

```python
#Remove duplicate numbers in a list
list1=[]
list2=[]

for x in age_tenants:
    if x not in list1:
        list1.append(x)
    elif x not in list2:
        list2.append(x)

print("The duplicate list is",list2)
del list2
print("The list without duplicacy is : ",list1)
```
