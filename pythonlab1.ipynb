```python
#Write a python program to encrypt a given string
message = input("Enter a string")
msg=[*message]
n= int(input("Enter the number to add to each character"))

for x in msg:
    ascii = ord(x)
    next_char = chr(ascii + n)
    print(next_char)
```

```python
#In ascending order by first numeric value
t_list=[ ('main st.', 4, 4000), ('elm st.', 1, 1200), ('pine st.', 2, 1600)]

def Sort(t_list): 
      t_list.sort(key = lambda a: a[1]) 
      return t_list 
    
   
print("In ascending order by first numeric value:" ,Sort(t_list))
```

```python
#In descending order by second numeric value
def Sort(t_list): 
      t_list.sort(key = lambda a: a[2],reverse=True) 
      return t_list 
    
   
print("In descending order by second numeric value:" ,Sort(t_list))
```

```python
#In alphabetical order of string value
def Sort(t_list): 
      t_list.sort(key = lambda a: a[0]) 
      return t_list 
    
   
print("In alphabetical order of string value:" ,Sort(t_list))
```
